Step:-5 — NumPy Simple Task

Example: Calculate mean score using NumPy

```python
math_scores = df['math score']
print("Average Math Score:", np.mean(math_scores))
```

```python
total_avg = np.mean([df['math score'],df['reading score'],df['writing score']])
print("Overall Average Score:", total_avg)
```

Step:-6 — Grouped Analysis (Pandas)

Example: Average score by gender

```python
# Average scores by gender for selected subjects
df.groupby('gender')[['math score','reading score','writing score']].mean()
```
