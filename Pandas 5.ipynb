```python
df1 = pd.DataFrame ({'Fruit' : ['banana', 'apple','orange', 'grape'], 'Wholesaler' : [100, 40 ,80, 60], 'market': [100,80,40,60]})

df1
```

```python
#merge two dataset with how

#--> method 

#left  --   Use keys from left object
#right --   Use keys from right object
#outer --   Use union of keys
#inner --   Use iintersection of keys
```

```python
df2 = pd.DataFrame ({'Fruit' : ['apple', 'banana', 'grape', 'kiwi'], 'Wholesaler' : [60,20,45,100]})
df2
```

```python
pd.merge(df1,df2, how='inner', on = 'Fruit')
```

```python
pd.merge(df1,df2, how='left', on='Fruit')
```

```python
pd.merge(df1,df2, how='outer', on='Fruit' )
```

```python
df3 = pd.DataFrame({'ID' : [1,2,3,4], 'Name' : ['a','b','c', 'd']})

df3
```

```python
df4 = pd.DataFrame({'ID' : [5,6,7,8], 'Name' : ['a','b','c', 'd']})

df4
```

```python
pd.merge(df3,df4, left_index=True, right_index=True) #Use Left & right indexes together
```

```python
pd.merge(df3,df4, left_index=True, right_index=True,suffixes=['priyraj','krish'])
```
